```python
income_groups = df['IncomeGroup'].value_counts()
income_groups.plot(kind='bar')
plt.title('Number of Countries in Each Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.show()
```

```python
regions = df['Region'].value_counts()
regions.plot(kind='bar')
plt.title('Number of Countries in Each Region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='IncomeGroup', kde=False)
plt.title('Distribution of Income Groups')
plt.xlabel('Income Group')
plt.ylabel('Frequency')
plt.show()
```
